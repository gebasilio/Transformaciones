# Correlación

## Definición

Para nuestro objetivo, la correlación cruzada (o a veces denominada "covarianza cruzada") es una medida de la similitud entre dos señales, frecuentemente usada para encontrar características relevantes en una señal desconocida por medio de la comparación con otra que sí se conoce.

El operador correlación entre dos señales discretas está definido como:

$$Corr(n)=\sum_{m=0}^{M-1}{f(m)g(m-n)}$$

## Ejemplo

```python
import numpy as np
from matplotlib import pyplot as plt
#Defino la señal desconocida:
x=np.random.randn(20000)

#Extraigo una parte de la señal desconocida, representará la señal conocida
y=np.random.randn(2000)
x[10000:12000]+=0.2*y
corr=np.correlate(x,y)
f, axarr = plt.subplots(3, sharex=True)
axarr[0].plot(x)
axarr[0].set_title('Sharing X axis')
axarr[1].plot(y)
axarr[2].plot(corr)
plt.show()

print("El máximo de la correlación está en:")
print(corr.argmax())
```

```python
# a^2 + b^2 =1 para que se mantenga la amplitud de las señales generadoras
#Defino la señal desconocida:
x=np.sin(np.linspace(0,2*np.pi,20000))

#Analizo la correlación con la misma señal
y=np.cos(np.linspace(0,2*np.pi,20000))

#a puede valer entre 1 y -1
a=-1
b=np.sqrt(1-a**2)

fas=a*x+b*y

corr=np.correlate(x,y)
f, axarr = plt.subplots(3, sharex=True)
axarr[0].plot(x)
axarr[0].set_title('Sharing X axis')
axarr[1].plot(y)
axarr[2].plot(fas)
plt.show()
```

## Representación de una señal periódica en función de senos y cosenos fijos de distintas frecuencias

```python
from scipy import signal
#genero una señal cuadrada
N=20000
x=signal.square(np.linspace(0.5*np.pi,20.5*np.pi,N), duty=0.5)

#Analizo la correlación con senos y cosenos
cos=np.cos(np.linspace(0,20*np.pi,N))
sin=np.sin(np.linspace(0,20*np.pi,N))
cos1=np.cos(np.linspace(0,40*np.pi,N))
sin1=np.sin(np.linspace(0,40*np.pi,N))
cos2=np.cos(np.linspace(0,60*np.pi,N))
sin2=np.sin(np.linspace(0,60*np.pi,N))
cos3=np.cos(np.linspace(0,80*np.pi,N))
sin3=np.sin(np.linspace(0,80*np.pi,N))
cos4=np.cos(np.linspace(0,100*np.pi,N))
sin4=np.sin(np.linspace(0,100*np.pi,N))
cos5=np.cos(np.linspace(0,120*np.pi,N))
sin5=np.sin(np.linspace(0,120*np.pi,N))
cos6=np.cos(np.linspace(0,140*np.pi,N))
sin6=np.sin(np.linspace(0,140*np.pi,N))
cos7=np.cos(np.linspace(0,160*np.pi,N))
sin7=np.sin(np.linspace(0,160*np.pi,N))

f, axarr = plt.subplots(3, sharex=True)
axarr[0].plot(x)
axarr[0].set_title('Sharing X axis')
axarr[1].plot(cos)
axarr[2].plot(sin)
plt.show()
b0=2*np.dot(x,sin)/N
b1=2*np.dot(x,sin1)/N
b2=2*np.dot(x,sin2)/N
b3=2*np.dot(x,sin3)/N
b4=2*np.dot(x,sin4)/N
b5=2*np.dot(x,sin5)/N
b6=2*np.dot(x,sin6)/N
b7=2*np.dot(x,sin7)/N
a0=2*np.dot(x,cos)/N
a1=2*np.dot(x,cos1)/N
a2=2*np.dot(x,cos2)/N
a3=2*np.dot(x,cos3)/N
a4=2*np.dot(x,cos4)/N
a5=2*np.dot(x,cos5)/N
a6=2*np.dot(x,cos6)/N
a7=2*np.dot(x,cos7)/N

print("La presencia de seno de f0,f1,f2 y f3 en la señal x es:")
print(b0)
print(b1)
print(b2)
print(b3)
print(b4)
print(b5)
print(b6)
print(b7)
print("La presencia de coseno de f0 en la señal x es:")
print(a0)
print(a1)
print(a2)
print(a3)
print(a4)
print(a5)
print(a6)
print(a7)
```

```python
rec=a0*cos+a1*cos1+a2*cos2+a3*cos3+a4*cos4+a5*cos5+a6*cos6+a7*cos7+b0*sin+b1*sin1+b2*sin2+b3*sin3+b4*sin4+b5*sin5+b6*sin6+b7*sin7
f, axarr = plt.subplots(3, sharex=True)
axarr[0].plot(rec)
axarr[0].set_title('Recuperación con 7 armónicos')
axarr[1].plot(cos)
axarr[2].plot(sin)
plt.show()
rec=a0*cos+a1*cos1+a2*cos2+a3*cos3+b0*sin+b1*sin1+b2*sin2+b3*sin3
f, axarr = plt.subplots(3, sharex=True)
axarr[0].plot(rec)
axarr[0].set_title('Recuperación con 3 armónicos')
axarr[1].plot(cos)
axarr[2].plot(sin)
plt.show()
```

# Transformada de Fourier

La transformada de Fourier, denominada así por Joseph Fourier, es una transformación matemática empleada para transformar señales entre el dominio del tiempo y el dominio de la frecuencia, que tiene muchas aplicaciones en la física y la ingeniería. Es reversible, siendo capaz de transformarse en cualquiera de los dominios al otro. El propio término se refiere tanto a la operación de transformación como a la función que produce.

En el caso de una función periódica en el tiempo (por ejemplo, un sonido musical continuo pero no necesariamente sinusoidal), la transformada de Fourier se puede simplificar para el cálculo de un conjunto discreto de amplitudes complejas, llamado coeficientes de las series de Fourier. Ellos representan el espectro de frecuencia de la señal del dominio-tiempo original.

# DFT/FFT

En matemáticas, la transformada discreta de Fourier o DFT (del inglés, discrete Fourier transform) es un tipo de transformada discreta utilizada en el análisis de Fourier. Transforma una función matemática en otra, obteniendo una representación en el dominio de la frecuencia, siendo la función original una función en el dominio del tiempo. Pero la DFT requiere que la función de entrada sea una secuencia discreta y de duración finita. Dichas secuencias se suelen generar a partir del muestreo de una función continua, como puede ser la voz humana. 

Utilizar la DFT implica que el segmento que se analiza es un único período de una señal periódica que se extiende de forma infinita; si esto no se cumple, se debe utilizar una ventana para reducir los espurios del espectro.

Se dice que la DFT es una transformada de Fourier para análisis de señales de tiempo discreto y dominio finito. 

La entrada de la DFT es una secuencia finita de números reales o complejos, de modo que es ideal para procesar información almacenada en soportes digitales. 

Un factor muy importante para este tipo de aplicaciones es que la DFT puede ser calculada de forma eficiente en la práctica utilizando el algoritmo de la transformada rápida de Fourier o FFT (Fast Fourier Transform).

Los algoritmos FFT se utilizan tan habitualmente para calcular DFTs que el término "FFT" muchas veces se utiliza en lugar de "DFT" en lenguaje coloquial. Formalmente, hay una diferencia clara: "DFT" hace alusión a una transformación o función matemática, independientemente de cómo se calcule, mientras que "FFT" se refiere a una familia específica de algoritmos para calcular DFTs.
